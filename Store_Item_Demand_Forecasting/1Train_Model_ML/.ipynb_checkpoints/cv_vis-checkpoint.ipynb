```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import datetime
from sklearn.model_selection import train_test_split
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

#pandasï¼š
pd.set_option('display.max_rows', 5000)
pd.set_option('display.max_columns', 5000)
pd.set_option('display.width', 1000)
```

### Lasso

```python
cv1 = pd.read_csv('cv1_results.csv').drop('Unnamed: 0',1)
cv1.head()
```

```python
cv1.index = [str(x) for x in cv1['param_alpha']]
```

```python
plt.figure(figsize = (6,4))
plt.plot(cv1.index, cv1['split0_test_score'])
plt.plot(cv1.index, cv1['split1_test_score'])
plt.plot(cv1.index, cv1['split2_test_score'])
#plt.plot(cv1.index, cv1['mean_test_score'])
plt.ylabel('R2 Score')
plt.xlabel('Alpha')
plt.axvline(x='0.1',color='black',linestyle='--')
plt.ylim([0.4,1])
plt.legend(loc='lower left')
plt.show()
```

### knn

```python
cv2 = pd.read_csv('cv2_results.csv').drop('Unnamed: 0',1)
cv2.head()
cv2.index = [str(x) for x in cv2['param_n_neighbors']]
```

```python
plt.figure(figsize = (6,4))
plt.plot(cv2.index, cv2['split0_test_score'])
plt.plot(cv2.index, cv2['split1_test_score'])
plt.plot(cv2.index, cv2['split2_test_score'])
#plt.plot(cv1.index, cv1['mean_test_score'])
plt.ylabel('R2 Score')
plt.xlabel('N Neighbours')
plt.axvline(x='15',color='black',linestyle='--')
plt.ylim([0.8,0.95])
plt.legend(loc='lower left')
plt.show()
```
