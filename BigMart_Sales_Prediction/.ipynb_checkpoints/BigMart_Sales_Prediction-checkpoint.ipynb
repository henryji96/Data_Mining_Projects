### 3.1 parse Item_Identifier
FDA15 - FD

```python
print(df_all['Item_Identifier'].nunique())

df_all['Item_Type2'] = df_train['Item_Identifier'].\
                                     apply(lambda x: x[:2]).\
                                     map({'FD':'Food','NC':'Non-Consumable','DR':'Drinks'})

print(df_all['Item_Type2'].value_counts())
```

### 3.2 Outlet Type is Supermarket or Grocery 

```python
df_all['Outlet_Type'].unique()
```

```python
def isSupermarket(Outlet_Type):
    if Outlet_Type.startswith('Supermarket'):
        return 1
    else:
        return 0
df_all['isSupermarket'] = df_all['Outlet_Type'].map(isSupermarket)
```

### 3.3 Compute Aggregation Features
by Outlet_Identifier, Item_MRP mean and set

```python
aggregation = df_all.groupby('Outlet_Identifier')['Item_MRP'].agg(['mean','std']).reset_index()\
                 .rename(columns={
                     'mean': 'By_Outlet_Identifier_MRP_Mean',
                     'std': 'By_Outlet_Identifier_MRP_Std'
})
print(aggregation.shape)
aggregation.head(2)
```

```python
df_all = df_all.merge(aggregation,how='left',on='Outlet_Identifier')
df_all.head(2)
```

### 3.4 Encode Ordered Factor and 2 Class Factor

```python
df_all['Item_Fat_Content'].unique()
```

# 4. Train Model

```python
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
```

### 4.1 train benchmark model

```python
lr = LinearRegression()
lr.fit(trainX, trainY)
trainY_pred = lr.predict(trainX)
rmse(trainY_pred, trainY)
```

```python
lasso = Lasso(alpha=1)
lasso.fit(trainX, trainY)
trainY_pred = lasso.predict(trainX)
rmse(trainY_pred, trainY)
```

```python
dt = DecisionTreeRegressor(min_samples_split=40,
                            min_samples_leaf=5,
                           max_features = 'auto')
dt.fit(trainX, trainY)
trainY_pred = dt.predict(trainX)
rmse(trainY_pred, trainY)
```

```python
rf = RandomForestRegressor(n_estimators=1500,
                            max_depth = 30,
                            max_features = 'sqrt',
                            min_samples_split = 10)
rf.fit(trainX, trainY)
trainY_pred = rf.predict(trainX)
rmse(trainY_pred, trainY)
```

```python
gbdt = GradientBoostingRegressor(n_estimators=1000, 
                                learning_rate=0.1,
                                min_samples_split = 10,
                                subsample = 0.4, 
                                max_leaf_nodes = 8,
                                max_features = 'sqrt')
gbdt.fit(trainX, trainY)
trainY_pred = gbdt.predict(trainX)
rmse(trainY_pred, trainY)
```
